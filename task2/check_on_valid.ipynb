```python
import pandas as pd

df = pd.read_json('dataset.json')
df.describe(include='all')
```

```python

keys = list()
for row in (set(dict(row).keys()) for row in df.params):
    keys.extend(list(row))
keys = list(set(keys))
```

```python
df1 = pd.DataFrame(columns=list(df.columns[:-1])+keys)
df1
```

```python
for i, row in df.iterrows():
    # buffer_params = pd.DataFrame({param[0]:[param[1]] for param in row['params'].items()})
    buffer_params = pd.Series(row['params'])
    df1.loc[i] = pd.concat([buffer_params,row[:-1]])
df1
```
